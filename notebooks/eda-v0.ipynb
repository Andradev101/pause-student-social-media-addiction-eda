```python
import os
import kagglehub
import pandas as pd 

# Define download path
os.environ["KAGGLEHUB_CACHE"] = "../datasets"

# Download dataset
path = kagglehub.dataset_download("adilshamim8/social-media-addiction-vs-relationships")
csv_file = os.path.join(path, "Students Social Media Addiction.csv")
df = pd.read_csv(csv_file)
```

```python
print(df.describe())
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# df.hist(edgecolor='black', linewidth=1, figsize=(12, 8))
# plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(
    data=df, 
    x='Avg_Daily_Usage_Hours', 
    kde=True, 
    color='skyblue',
    hue="Gender",
    multiple="stack"
)
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(
    data=df, 
    x='Avg_Daily_Usage_Hours', 
    kde=True, 
    color='skyblue',
    hue="Academic_Level",
    multiple="stack"
)
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(
    data=df, 
    x='Sleep_Hours_Per_Night', 
    kde=True, 
    color='skyblue',
    hue="Gender",
    multiple="stack"
)
plt.show()

sns.violinplot(data=df, x="Academic_Level", y="Age")
plt.show()
# sns.pairplot(df)
# plt.suptitle("Pairplot of Features", y=1.02)
# plt.show()

# plt.figure(figsize=(12, 6))
# sns.boxplot(data=df, width=0.5, fliersize=5)
# plt.title("Boxplot of Features")
# plt.show()

# plt.figure(figsize=(12, 6))
# sns.violinplot(x='Age', y='Addicted_Score', data=df)
# plt.title("Violin Plot of Sepal Length")
# plt.show()

# correlation_matrix = df.drop(columns = ['Age']).corr()
# plt.figure(figsize=(8, 6))
# sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
# plt.title("Correlation Matrix Heatmap")
# plt.show()

sns.lmplot(x='Age', y='Addicted_Score', data=df)
plt.title("Sepal Length vs Sepal Width with Regression Line")
plt.show()
```
